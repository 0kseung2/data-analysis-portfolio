# 전국 주요도시 시간대별 미세먼지 농도 변화 분석

## 주제 선정 이유

현재 미세먼지로 인해 발생한는 질병과 기타 피해들이 심각하다. 그렇기에 현재 전 국민이 미세먼지에 대해 경각심을 가지고 있는 상황이다.<br>

특히 봄철에 미세먼지가 심각하기에 봄철의 날짜 중 하루를 선정하고, 미세먼지 농도를 시간별로 측정함으로써 시간대별로 미세먼지 농도가 어떻게 변화하는지 시각적으로 파악할 수 있게 한다. 그리하여 시간대별로 생활습관을 어떻게 교정해 나가야 할 것인가에 대해 고찰하게끔 한다.<br> 

또한 지도시각화를 이용하여 본인이 사는 지역 혹은 타지역의 미세먼지 농도를 알아볼 수 있음으로써 지역별로 미세먼지 농도가 얼마나 차이가 나는지 알아볼 수 있고, 그를 통해 미세먼지의 심각성을 느끼고 대책을 세울 수 있을 것이다.<br>

더불어 미세먼지 농도를 시각적으로 볼 수 있음으로써 현재 얼마나 환경문제가 심각한지 인식할 수 있도록 해준다.

## 가설 정의

봄철인 3월의 중순인 15일로 날짜를 선택하였으며, 지역과 시간대별로 미세먼지 농도를 분석한다.<br>

꺾은선 그래프와 같은 데이터 시각화와 지도 시각화를 통해서 지역별, 시간대별, 대기오염 농도 분석을 실시할 것이다.<br>

이때, 수도권과 같이 인원수가 많거나, 황사를 타고 미세먼지가 많이 날라오는 서쪽에 가까운 지역일수록 미세먼지 농도가 높을 것으로 예상된다.<br>

시간대별로는 사람들의 활동이 주로 일어나는 시간대인 낮시간대가 미세먼지 농도가 가장 높을 것으로 예상되고, 전후로 하여 미세먼지 농도가 상승, 하락 하는 추세를 나타낼 것으로 예상된다.<br>

## 분석을 위한 데이터의 가공

1. 미세먼지 농도 값 변환
- loc를 이용하여 미세먼지 농도값이 측정되지 않아 -999로 되어 있는 값을 0으로 변환하였다.

2. 측정일시와 미세먼지 농도 값을 데이터 프레임으로 가져옴
- loc와 str.contains를 이용하여 시도와 군구(혹은 측정소명)를 기준으로 하여 3월 15일의 시간대별 미세먼지 농도 값을 가져왔다.

3. 지도상에 만들어지는 그래프의 X축 값들을 연속성있게 만듬
- 지도상에 그래프를 만들 때, X축 값들(측정 시간)이 기존 데이터값들로는 불연속적이게 나왔다.(예를 들어 2021-03-15 3:00이 2021-03-15 13:00보다 큰 값으로 나타났다. 시간에서 앞의 3이 1보다 크다고 인식하였기 때문) 그리하여 change_time 함수를 만들어 앞의 자릿수가 한자리 수인 시간대들의 앞에 0을 붙여주며 데이터 프레임의 값을 변환시켜 X축의 시간값들이 연속으로 구성되게 하였다.

## 분석 결과 도출

전국 시간대별 대기오염물질 농도 중 주요도시별 특정 날짜에서의 시간대별 미세먼지 농도 데이터를 획득함으로써 그래프를 그리는 시각화를 통해 지역별 미세먼지 농도 차이와, 시간대별 미세먼지 농도의 변화양상을 구하였다.

또한 지역별 미세먼지 농도 데이터프레임으로부터 미세먼지 농도의 평균값을 구하였다.<br> 
지도 시각화를 통해 지역별로 다른 미세먼지 농도 평균값을 원의 직경으로 설정하여 지역별로 평균값의 크기에 따라 원의 크기가 다르게 나타남을 보임으로써 지역별로 미세먼지 농도의 차이가 직관적으로 보이게 하였다.<br>
더불어 지도 상에서도 그래프를 그려보임으로써 지도 시각화라는 하나의 도구 속에서 그래프도 얻을 수 있고, 직관적으로 데이터 측정치를 파악할수 있음을 보였다. 이로부터 하나의 도구로 한번에 여러 정보들을 시각적으로 획득할 수 있음을 보였다.

## Python 소스코드 원본

```python
import folium
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import altair as alt
```

```python
csv = pd.read_csv("환경부 국립환경과학원_대기오염도 확정자료_20211231.csv", encoding='cp949')
csv.loc[csv["미세먼지"] == -999, "미세먼지"] = 0 # 미세먼지 농도가 -999인 값은 0으로 변경하여 분석을 진행하였다.
print(csv)
```

```python
#그래프 상의 축 제목 한글 폰트로 지정
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)

#그래프 크기 키우기 (x축의 측정시간이 겹치는 부분 줄이는 것)
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
#강원, 경기, 경남, 경북, 광주, 대구, 대전, 부산 순으로 각각 도/시별 하나의 군구/읍/동을 택하여 분석을 진행하였다.
#시도와 군구를 선택하고, 그에 따른 미세먼지 농도와 측정일시를 데이터 프레임으로 만들었다.

#강원
csv_gangwon = csv.loc[(csv["시도"]=="강원")&(csv["군구"]=="인제군"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_gangwon = csv_gangwon[csv_gangwon['측정일시'].str.contains('2021-03-15')]

#경기
csv_gyeonggi = csv.loc[(csv["시도"]=="경기")&(csv["측정소명"]=="양평읍"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_gyeonggi = csv_gyeonggi[csv_gyeonggi['측정일시'].str.contains('2021-03-15')]

#경남
csv_gyeongnam = csv.loc[(csv["시도"]=="경남")&(csv["군구"]=="의령군"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_gyeongnam = csv_gyeongnam[csv_gyeongnam['측정일시'].str.contains('2021-03-15')]

#경북
csv_gyeongbuk = csv.loc[(csv["시도"]=="경북")&(csv["군구"]=="안동시"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_gyeongbuk = csv_gyeongbuk[csv_gyeongbuk['측정일시'].str.contains('2021-03-15')]

#광주
csv_gwangju = csv.loc[(csv["시도"]=="광주")&(csv["측정소명"]=="치평동"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_gwangju = csv_gwangju[csv_gwangju['측정일시'].str.contains('2021-03-15')]

#대구
csv_daegu = csv.loc[(csv["시도"]=="대구")&(csv["군구"]=="중구"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_daegu = csv_daegu[csv_daegu['측정일시'].str.contains('2021-03-15')]

#대전
csv_daejeon = csv.loc[(csv["시도"]=="대전")&(csv["측정소명"]=="둔산동"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_daejeon = csv_daejeon[csv_daejeon['측정일시'].str.contains('2021-03-15')]

#부산
csv_busan = csv.loc[(csv["시도"]=="부산")&(csv["군구"]=="동래구"), ["시도", "군구", "측정일시", "미세먼지"]]
dataframe_busan = csv_busan[csv_busan['측정일시'].str.contains('2021-03-15')]

#도시별 측정일시, 미세먼지 농도 데이터 프레임 만들기
dataframe_cities = [dataframe_gangwon, dataframe_gyeonggi, dataframe_gyeongnam, dataframe_gyeongbuk, dataframe_gwangju, 
                    dataframe_daegu, dataframe_daejeon, dataframe_busan]

#display dataframe
for dataframe_city in dataframe_cities:
    display(dataframe_city)
```

```python
#folium으로 지도 시각화 한 것 위에 그래프를 만드는데, 원래 데이터에 있던 시간대로 하면 
#시간 순서가 바뀌어서 데이터를 수정하여 진행하였다.
def change_time(values):
    if values == '2021-03-15 0:00':
        return '00:00'
    elif values == '2021-03-15 1:00':
        return '01:00'
    elif values == '2021-03-15 2:00':
        return '02:00'
    elif values == '2021-03-15 3:00':
        return '03:00'
    elif values == '2021-03-15 4:00':
        return '04:00'
    elif values == '2021-03-15 5:00':
        return '05:00'
    elif values == '2021-03-15 6:00':
        return '06:00'
    elif values == '2021-03-15 7:00':
        return '07:00'
    elif values == '2021-03-15 8:00':
        return '08:00'
    elif values == '2021-03-15 9:00':
        return '09:00'
    elif values == '2021-03-15 10:00':
        return '10:00'
    elif values == '2021-03-15 11:00':
        return '11:00'
    elif values == '2021-03-15 12:00':
        return '12:00'
    elif values == '2021-03-15 13:00':
        return '13:00'
    elif values == '2021-03-15 14:00':
        return '14:00'
    elif values == '2021-03-15 15:00':
        return '15:00'
    elif values == '2021-03-15 16:00':
        return '16:00'
    elif values == '2021-03-15 17:00':
        return '17:00'
    elif values == '2021-03-15 18:00':
        return '18:00'
    elif values == '2021-03-15 19:00':
        return '19:00'
    elif values == '2021-03-15 20:00':
        return '20:00'
    elif values == '2021-03-15 21:00':
        return '21:00'
    elif values == '2021-03-15 22:00':
        return '22:00'
    else:
        return '23:00'

#지역별 데이터 프레임에서 측정일시 값을 함수를 사용하여 위의 조건에 맞게 변환하였다.
dataframe_gangwon['측정일시'] = dataframe_gangwon['측정일시'].apply(change_time)
dataframe_gyeonggi['측정일시'] = dataframe_gyeonggi['측정일시'].apply(change_time)
dataframe_gyeongnam['측정일시'] = dataframe_gyeongnam['측정일시'].apply(change_time)
dataframe_gyeongbuk['측정일시'] = dataframe_gyeongbuk['측정일시'].apply(change_time)
dataframe_gwangju['측정일시'] = dataframe_gwangju['측정일시'].apply(change_time)
dataframe_daegu['측정일시'] = dataframe_daegu['측정일시'].apply(change_time)
dataframe_daejeon['측정일시'] = dataframe_daejeon['측정일시'].apply(change_time)
dataframe_busan['측정일시'] = dataframe_busan['측정일시'].apply(change_time)


#데이터 프레임의 측정일시 값들이 수정되어 있음을 확인할 수 있다.
display(dataframe_gangwon)
display(dataframe_gyeonggi)
display(dataframe_gyeongnam)
display(dataframe_gyeongbuk)
display(dataframe_gwangju)
display(dataframe_daegu)
display(dataframe_daejeon)
display(dataframe_busan)
```
